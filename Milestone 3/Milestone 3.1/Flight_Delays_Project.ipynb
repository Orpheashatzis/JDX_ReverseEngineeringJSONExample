```python
# Showing missing values
missing_values = df.isnull().sum()
print("Missing Values :")
print(missing_values)
```

```python
# Remove rows and columns with missing values (just in case)
df.dropna(inplace=True) 
df.dropna(axis=1, inplace=True) 
```

```python
# Getting a list of airport codes
airport_code = df['Origin'].unique()
airport_code
```

```python
# Getting the 10 busiest airports in USA for 2022
buisiest_airports = df['Origin'].value_counts().nlargest(10)
buisiest_airports
```

```python
# Bar plot to visualise top 10 buisiest airports
plt.figure(figsize=(10,7))
sns.barplot(x=buisiest_airports.index, y=buisiest_airports.values)
plt.title('Top 10 Buisiest Airports in USA 2022')
plt.xlabel('Airport')
plt.ylabel('Number of Flights')
plt.tight_layout()
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Getting the 10 least busy airports in USA for 2022
least_busy_airports = df['Origin'].value_counts().nsmallest(10)
least_busy_airports
```

```python
# Bar plot to visualise top 10 least busy airports
plt.figure(figsize=(10,7))
sns.barplot(x=least_busy_airports.index, y=least_busy_airports.values)
plt.title('Top 10 Least Busy Airports in USA 2022')
plt.xlabel('Airport')
plt.ylabel('Number of Flights')
plt.tight_layout()
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Convert CRSDepTime to hours
df['DepHour'] = df['CRSDepTime'] // 100
```

```python
# Calculate avergage delay by hour
average_delay_by_hour = df.groupby('DepHour')['DepDelay'].mean().reset_index()
# Bar plot to visualize average flight delays by Time of Day
plt.figure(figsize=(12, 6))
sns.barplot(data=average_delay_by_hour, x='DepHour', y='DepDelay')
plt.title('Average Flight Delays by Time of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Average Delay (mins)')
plt.xticks(rotation=45)  
plt.show()
```

```python
# Plot to visualise distribution of flights delays less than 30 mins
df.query("DepDelayMinutes < 30")["DepDelayMinutes"].plot(kind="hist", bins=30, title="Distribution of Flight Delays < 30 Min")
plt.show()
```

```python
# Most of the flight delays seem to be between 0-1 mins
```

```python
# Plot to visualise distribution of flight delays more than 30 mins
df.query("DepDelayMinutes > 1 and DepDelayMinutes < 61")["DepDelayMinutes"].plot(kind="hist", bins=30, title="Distribution of Flight Delays < 30 Min")
plt.show()
```

```python
# Grouping of delays
df["DelayGroup"] = None
df.loc[df["DepDelayMinutes"] == 0, "DelayGroup"] = "OnTime_Early"
df.loc[(df["DepDelayMinutes"] > 0) & (df["DepDelayMinutes"] <= 15), "DelayGroup"] = "Small_Delay"
df.loc[(df["DepDelayMinutes"] > 15) & (df["DepDelayMinutes"] <= 45), "DelayGroup"] = "Medium_Delay"
df.loc[df["DepDelayMinutes"] > 45, "DelayGroup"] = "Large_Delay"
df.loc[df["Cancelled"], "DelayGroup"] = "Cancelled"
```

```python
# PLot to visualise type of delay
df["DelayGroup"].value_counts(ascending=True).plot(kind="barh", figsize=(10, 5), title="Flight Results 2022")
plt.show()
```

```python
# most flights are on time
# if there is a delay it is most likelt a small delay <= 15 mins
# however tendancy of medium or large delay are on similar frequencies
```

```python
plt.scatter(df['DepDelay'], df['ArrDelay'], alpha=0.5)
plt.title('Departure Delay vs Arrival Delay')
plt.xlabel('Departure Delay (minutes)')
plt.ylabel('Arrival Delay (minutes)')
plt.grid(True)
plt.show()
df[['DepDelay','ArrDelay']].corr()
```

```python
numerical_features = df[['ArrDel15','DepDel15', 'WheelsOff' , 'WheelsOn','ArrDelay','DistanceGroup','CRSDepTime', 'CRSArrTime', 'AirTime', 'Distance','DepHour', 'TaxiIn','TaxiOut',]]
correlation_matrix = numerical_features.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
# Convert FlightDate to datetime
df['FlightDate'] = pd.to_datetime(df['FlightDate'])
```

```python
# Adding time base features
# Hour done above 'DepHour'
df['day_of_week'] = df['FlightDate'].dt.day_name()
df['month'] =  df['FlightDate'].dt.month
```

```python
# Plot to visualise delays by month
plt.figure(figsize=(10,6))
sns.barplot(x='month', y='DepDelayMinutes', data= df , errorbar=None)
plt.title('Average Departure Delay by Month')
plt.show()
```

```python
# Can see that month does not really affect delays
```

```python
df_agg = df.groupby("Month")["DelayGroup"].value_counts(normalize=True).unstack() * 100
col_order = ["OnTime_Early", "Small_Delay", "Medium_Delay", "Large_Delay"]
df_agg[col_order].style.background_gradient(cmap="Blues")
```

```python
# Plot to visualise delays by day of the week
plt.figure(figsize=(12, 6))
sns.barplot(x='day_of_week', y='DepDelayMinutes', data=df, 
            order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], errorbar=None)
plt.title('Average Departure Delay by Day of the Week')
plt.show()
```

```python
#Friday has the most delays and then the weekends. Tuesday and Wednesday have the shortest delay.
#Delay time rises from Tuesday until it reaches max (Friday) then starts to decrease again.
```

```python
# which airline has the most flights
fig, ax = plt.subplots(figsize=(10, 7))
airlines_ordered = (df["Airline"].value_counts(ascending=True) / 100_000).plot(kind="barh", ax=ax, width=1, edgecolor="black")
ax.set_title("Number of Flights in Dataset")
ax.set_xlabel("Flights (100k)")
```

```python
#Since weekends and fridays had the greatest time delays we dedcided to sort our data into the three sopecified days to further examine the data and come to conclusions
```

```python
# Creating a new dataframe for only weekend and friday flights
df_weekend = df[df['day_of_week'].isin(['Friday', 'Saturday', 'Sunday'])]
df_weekend.head()
```

```python
df_weekend.shape
```

```python
# Calculate avergage delay by hour
average_delay_by_hour_weekend = df_weekend.groupby('DepHour')['DepDelay'].mean().reset_index()
# Bar plot to visualize average flight delays by Time of Day
plt.figure(figsize=(12, 6))
sns.barplot(data=average_delay_by_hour_weekend, x='DepHour', y='DepDelay')
plt.title('Average Flight Delays by Time of Day On weekends/Fridays')
plt.xlabel('Hour of Day')
plt.ylabel('Average Delay (mins)')
plt.xticks(rotation=45)  
plt.show()
```

```python
# Creating a new dataframe for only Mon, tue, wed, thu
df_not_weekend = df[df['day_of_week'].isin(['Monday', 'Tuesday', 'Wednesday','Thursday'])]
df_not_weekend.head()
```

```python
# Calculate avergage delay by hour
average_delay_by_hour_not_weekend = df_not_weekend.groupby('DepHour')['DepDelay'].mean().reset_index()
# Bar plot to visualize average flight delays by Time of Day
plt.figure(figsize=(12, 6))
sns.barplot(data=average_delay_by_hour_not_weekend, x='DepHour', y='DepDelay')
plt.title('Average Flight Delays by Time of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Average Delay (mins)')
plt.xticks(rotation=45)  
plt.show()
```
